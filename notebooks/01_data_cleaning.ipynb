### Import Required Libraries
The following libraries are used for data manipulation and preliminary analysis.

```python
import pandas as pd
```

### Load Dataset
The dataset is loaded into a Pandas DataFrame for further inspection and preprocessing.

```python
df = pd.read_csv("../data/raw/bank_fraud.csv")
```

### Initial Dataset Inspection
Before making any modifications, let's understand the overall structure of the dataset, including the number of rows, columns, data types, and a preview of the records.

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

#### Observation
1. One column (fraud_type) contains missing values.
2. Two columns (transaction_time and transaction_date) are of object datatypes. <br>
Appropriate handling will be performed in the future sections.

### Correcting Datatypes of Columns
transaction_time and transaction_date are of object. <br>
**Solution** - Lets combine the two and set their datatypes to datetime64

```python
df["transaction_datetime"] = pd.to_datetime(
    df["transaction_date"] + " " + df["transaction_time"]
)
```

```python
df.info()
```

```python
df.drop(["transaction_time", "transaction_date"], axis = 1, inplace = True)
```

### Handling NaN values
The missing values are just non-frauds which will be dropped to keep only the frauds in cleaned dataset.

```python
df[df["fraud_type"].isnull()]["is_fraud"].unique()
```

```python
df.dropna(subset = ["fraud_type"], inplace = True)
```

```python
df.isnull().sum()
```

### Looking for Duplicate data

```python
df.duplicated().sum()
```

#### Observation
There are no duplicate rows.

## **Save Cleaned Dataset**
After completing the cleaning process, the cleaned dataset is exported for use in subsequent notebooks.

```python
df.to_csv("../data/processed/bank_fraud_cleaned.csv")
```
