### Correcting Datatypes of Columns
transaction_time and transaction_date are of object. <br>
**Solution** - Lets combine the two and set their datatypes to datetime64

```python
df["transaction_datetime"] = pd.to_datetime(
    df["transaction_date"] + " " + df["transaction_time"]
)
```

```python
df.info()
```

```python
df.drop(["transaction_time", "transaction_date"], axis = 1, inplace = True)
```

### Handling NaN values
The column fraud_type has NaN values

```python
df["fraud_type"].isnull().sum()
```

```python
df[df["is_fraud"] == 0]["fraud_type"].isnull().sum()
```

```python
df[df["is_fraud"] == 1]["fraud_type"].isnull().sum()
```

```python
df.fillna(
    {
        "fraud_type": "No Fraud"
    }, inplace = True
)
```

```python
df.isnull().sum()
```

**Observation**: The fraud_type column contained missing values for transactions that were not fraudulent. Since a fraud type is only applicable when a transaction is classified as fraud, these missing values represent the absence of fraud rather than incomplete data. Therefore, the missing values were replaced with "No Fraud" preserving all observations.

### Looking for Duplicate data

```python
df.duplicated().sum()
```

**Observation**: There are no duplicate rows.
